```python
import pandas as pd
import numpy as np
import csv 
import shutil
import os, os.path
```

```python
f = open('hemorrhage_diagnosis.csv')
```

```python
reader = csv.reader(f)
```

```python
PatientNumber = []
No_Hemmorhage = []
SliceNumber = []
```

```python
for row in reader:
    PatientNumber.append(row[0])
    No_Hemmorhage.append(row[7])
PatientNumber.remove(PatientNumber[0])
No_Hemmorhage.remove(No_Hemmorhage[0])
```

```python
start = 49
index = []
counter = 1
```

```python
for pt in PatientNumber:
    if pt==str(start):
        index.append(pt+str(counter))
        counter += 1 
    else: 
        index.append(pt+str(1))
        counter = 1 
        counter += 1
        start += 1
       
```
